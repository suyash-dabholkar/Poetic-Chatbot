**STEP 4 - ADDING A STREAM FUNCTION TO SPLIT THE TEXT INTO WORDS**

```python
def stream(text):
  print("Bot :", end = "")
  words = list(text)
  for word in words:
    print(word , end="", flush=True)
    time.sleep(0.1)
  print()
```

**STEP 5 - CONNECTING GEMINI API**

```python
try:
  key = userdata.get("geminiapichatbotkey")
  client = genai.Client(api_key=key)
except KeyError:
  print(f"Error with Gemini API key.\n error message:\n{KeyError}")
chat = client.chats.create(model="gemini-2.0-flash")
```

**STEP 6 - CREATING FUNCTION TO TAKE INPUT AND TALK WITH GEMINI**

```python
def ask_gemini(chat_input):
  try:
    prompt = f"You are a poetic chatbot.Respond every question in a poetic way.Keep it short and unique an creative. Her's the question:{user_input}"
    response = chat.send_message(prompt)
    return response.text
  except Exception as e:
    return f"Error with Gemini API call.: {e}"
```

**STEP 7 - FINALLY MAKING A CHATBOT LOOP**

```python
print("Welcome to your Gemini Chatbot!Type 'quit' to exit.")
while True:
  user_input = input("You: ")
  if user_input.lower() == "quit":
    print("Bot:Bye for now!")
    break
  response = ask_gemini(user_input)
  stream(response)
```
